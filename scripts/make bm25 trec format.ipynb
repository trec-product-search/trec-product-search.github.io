```python
import json
q2data = {}
with open('dev.jsonl','r') as f:
    for l in f:
        data = json.loads(l)
        q2data[data['query']] = data
        
```

```python
q2data[data['query']]
```

```python
files = ['esci-desconly.txt',
         'esci-title-text-desconly.txt',
         'esci-title-textonly.txt',
         'esci-titleonly.txt',
         'esci-textonly.txt',]
```

```python
for file in files:
    ofile = file[:-3] + 'trec'
    print(ofile)
    with open('ranking/'+file,'r') as f:
        with open('ranking/'+ofile,'w') as w:
            for l in f:
                l = l.strip().split('\t')
                if l[0] in q2data:
                    qid = q2data[l[0]]['query_id']
                    w.write("{}\tQ0\t{}\t{}\t{}\tteam\n".format(qid, l[2],l[1],l[3]))
```
